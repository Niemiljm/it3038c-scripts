```python
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap
from matplotlib.ticker import MaxNLocator
import numpy as np
%matplotlib inline



```

```python
apl_price = [93.95, 112.15, 104.05, 144.85, 169.49] # Data from apple stock prices
ms_price = [39.01, 50.29, 57.05, 69.98, 94.39]      # Data from microsoft stock prices
year = [2014, 2015, 2016, 2017, 2018]               # Year data that I will use as the x axis
```

```python
plt.plot(year, apl_price, '--k',     # --k makes the plotted line dotted, and black. (-- = dotted line, k = black color)
           year, ms_price, 'c')      # c makes the line cyan!

plt.xlabel('Year')                   # x label
plt.ylabel('Stock Price')            # y label

# plt.axis([2013, 2019, 20, 180])    # If this is uncommented and run, then it zooms the graph out! It's pretty neat that the size of the graph is determined by the data you give and we can override it.

plt.show()                           # This is how you show the graph inline with the code
```

```python
fig_1 = plt.figure(1, figsize=(15, 5))                              # fig_1 is the ENTIRE thing below, not just one of the graphs. the figsize() value can be changed, the 15 being width, 5 being height.

apple_price = fig_1.add_subplot(121)                                # The (122) number contains 3 different value, each one resulting in a different looking graph!
apple_price.xaxis.set_major_locator(MaxNLocator(integer=True))      # I wanted to add this because it would get rid of the decimal in the year like seen above.

plt.xlabel('Year')                                                  # Simple x and y labels
plt.ylabel('Stock Price')

microsoft_price = fig_1.add_subplot(122)
microsoft_price.xaxis.set_major_locator(MaxNLocator(integer=True))

plt.xlabel('Year')
plt.ylabel('Stock Price')

apple_price.plot(year, apl_price)                                   # .plot plots the data in a line
microsoft_price.scatter(year, ms_price)                             # .scatter plots the data in a scatter

plt.show()
```

```python
fig_2, axes = plt.subplots(3,2,figsize=(20, 5))                    # This is an inline way of creating more than one graph, 3x2 size, with 20 width and 5 height.
axes[0,1].scatter(year, ms_price)                                  # uses the numpy package (axes) to determine where to plot the data. That's why there are empty graphs.
axes[1,0].plot(year, apl_price)
plt.show()
```

```python
# I wanted to learn some "advanced" uses of this software, so I followed a tutorial
# that is on matplotlib.org to get these cool color graphs. I changed quite a few of the values
# just to mess around with it and find out what each value does, I added my findings to the commented
# sections below!
x = np.arange(0, np.pi, 0.1)                     # Incremental x value
y = np.arange(0, 2 * np.pi, 0.1)                 # Incremental y value
X, Y = np.meshgrid(x, y)
Z = np.cos(X) * np.sin(Y) * 10                   # Calculations that give the patterns

colors = [(1, 90, 0), (0, 1, 0), (0, 0, 1)]      # The numbers are RGB references!
n_bins = [60, 2, 10, 100]                        # This kind of works like a resolution to the image
cmap_name = 'prettyColors'
fig, axs = plt.subplots(2, 2, figsize=(9, 10))

for n_bin, ax in zip(n_bins, axs.ravel()):
    
    # Creates the colormap and uses the information shown above to display them below
    cmap = LinearSegmentedColormap.from_list(cmap_name, colors, N=n_bin)
    im = ax.imshow(Z, origin='lower', cmap=cmap)
    
    # Sets a title and shows the n_bin value that was determined above
    ax.set_title("Pretty Colormap: %s" % n_bin)
    fig.colorbar(im, ax=ax)
```
